```python
import pandas as pd  # Importation de la bibliothèque pandas pour la manipulation et l'analyse des données sous forme de DataFrame
from sklearn.tree import DecisionTreeClassifier, plot_tree  # Importation de DecisionTreeClassifier pour créer un modèle d'arbre de décision, et plot_tree pour visualiser l'arbre
import matplotlib.pyplot as plt  # Importation de matplotlib pour la création de graphiques et de visualisations
import warnings  # Importation de warnings pour gérer les avertissements
import numpy as np  # Importation de NumPy pour la manipulation de tableaux et les calculs numériques
from sklearn.preprocessing import LabelEncoder  # Importation de LabelEncoder pour transformer les variables catégorielles en variables numériques
```

```python
# Chargement du fichier Excel 
data= pd.read_excel('C:/Users/asus/Downloads/car__evaluation.xlsx', header=None)
print(data)
```

```python
print('data',data)
# Définition des noms des colonnes pour le jeu de données
col_names = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'class']
# Assignation des noms de colonnes au DataFrame
data.columns = col_names
col_names
```

```python
# Suppression de la première ligne du DataFrame (index 0)
data = data.drop(index=0)

# Affichage du DataFrame après suppression de la première ligne
print(data)

# Vérification des valeurs nulles : affiche un DataFrame booléen (True pour les valeurs nulles)
print(data.isnull())

# Affiche le nombre total de valeurs nulles pour chaque colonne
print(data.isnull().sum())

```

```python
# Suppression des lignes où la colonne 'buying' contient des valeurs nulles
data = data.dropna(subset=['buying'])

# Suppression des lignes où la colonne 'safety' contient des valeurs nulles
data = data.dropna(subset=['safety'])

# Suppression des lignes où la colonne 'class' contient des valeurs nulles
data = data.dropna(subset=['class'])

# Affiche le nombre total de valeurs nulles par colonne après le nettoyage
print(data.isnull().sum())

```

```python
# Vérifie s'il existe des lignes en double dans le DataFrame
doublons_exist = data.duplicated().any()

# Affiche un message indiquant s'il y a des doublons dans le DataFrame
print("Y a-t-il des doublons dans le DataFrame ?", doublons_exist)

```

```python
# Supprime toutes les lignes en double du DataFrame
data = data.drop_duplicates()
```

```python
# Vérifie s'il reste des lignes en double dans le DataFrame
doublons_exist = data.duplicated().any()

# Affiche un message indiquant s'il y a encore des doublons dans le DataFrame
print("Y a-t-il des doublons dans le DataFrame ?", doublons_exist)

```

```python
# Séparation des caractéristiques (X) et de la cible (y)

# X contient toutes les colonnes sauf 'class' (caractéristiques)
X = data.drop(['class'], axis=1)

# y contient uniquement la colonne 'class' (cible)
y = data['class']
```

```python




# Initialiser le LabelEncoder
label_encoder = LabelEncoder()

# Appliquer LabelEncoder sur chaque colonne spécifiée
for col in ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety']:
    X[col] = label_encoder.fit_transform(X[col])  # Encode les valeurs de chaque colonne en entiers

# Afficher le DataFrame transformé
print(X)

```

```python
# Encoder la variable cible 'y' en valeurs numériques avec LabelEncoder
Y = label_encoder.fit_transform(y)

# Afficher la variable cible transformée
print(Y)
```

```python


# Initialisation du modèle d'Arbre de Décision avec certains paramètres
clf = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=0)

# Entraînement du modèle sur les caractéristiques X et la cible y
clf.fit(X, y)

# Visualisation de l'Arbre de Décision
plt.figure(figsize=(20, 10))  # Définir la taille de la figure pour une meilleure lisibilité
plot_tree(clf, feature_names=X.columns, class_names=True, filled=True)  # Dessin de l'arbre de décision
plt.title("Visualisation de l'Arbre de Décision")  # Titre de la figure
plt.show()  # Afficher le graphique
```
